```python
def init_params():
    w1 = np.random.rand(10,60) - 0.5
    b1 = np.random.rand(10,1) - 0.5
    
    w2 = np.random.rand(2,10) - 0.5
    b2 = np.random.rand(2,1) - 0.5
    
    return w1, b1, w2, b2
```

```python
def relu(z):
    return np.maximum(0,z)
```

```python
def softmax(z):
    return np.exp(z) / np.sum(np.exp(z))
```

```python
def forwardProp(w1, b1, w2, b2, X):
    z1 = w1.dot(X) + b1
    a1 = relu(z1)
    
    z2 = w2.dot(a1) + b2
    a2 = softmax(z2)
    
    return z1, a1, z2, a2
```

```python
def getReLuDerivative(Z):
    return Z > 0
```

```python
def backProp(z1, a1, z2, a2, w2, X, Y):
    m = Y.size
    oneHotY = oneHot(Y)
    dZ2 = a2 - oneHotY
    dW2 = 1/m * dZ2.dot(a1.transpose())
    db2 = 1/m * np.sum(dZ2)
    
    dZ1 = w2.transpose().dot(dZ2) * getReLuDerivative(z1)
    dW1 = 1/m * dZ1.dot(X.transpose())
    db1 = 1/m * np.sum(dZ1)
    
    return dW1, db1, dW2, db2
```

```python
def updateParams(dW1, db1, dW2, db2, w1, b1, w2, b2, alpha):
    w1 = w1 - alpha*dW1
    b1 = b1 - alpha*db1
    w2 = w2 - alpha*dW2
    b2 = b2 - alpha*db2
    
    return w1, b1, w2, b2
```

```python
def getPredictions(a2):
    return np.argmax(a2, 0)

def getAccuracy(predictions,Y):
    #print(predictions,Y)
    return np.sum(predictions == Y) / Y.size
```

```python
def gradientDescent(X, Y, epochs, lr):
    w1, b1, w2, b2 = init_params()
    
    for i in range(epochs):
        z1, a1, z2, a2 = forwardProp(w1, b1, w2, b2, X)
        dW1, db1, dW2, db2 = backProp(z1, a1, z2, a2, w2, X, Y)
        w1, b1, w2, b2 = updateParams(dW1, db1, dW2, db2, w1, b1, w2, b2, lr)
        
        if i%10 == 0:
            #print(a2)
            print(f'Iteration {i}')
            print('Accuracy:',getAccuracy(getPredictions(a2), maskY(Y)))
            
    return w1, b1, w2, b2
```

```python
def maskY(Y):
    masked = (Y == 'M')
    return masked.astype(int)
```

```python
def oneHot(Y):
    Y = maskY(Y)
    oneHotY = np.zeros((Y.size, Y.max()+1))
    oneHotY[np.arange(Y.size),Y] = 1
            
    return oneHotY.transpose()
```
